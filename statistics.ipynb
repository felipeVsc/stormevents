```python
import matplotlib.pyplot as plt

counts = details['EVENT_TYPE'].value_counts()[:5]
labels = counts.index.tolist()
values = counts.values.tolist()

plt.bar(labels, values)
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

df = location[['LATITUDE','LONGITUDE']]
df = df[df['LONGITUDE']<-50].sample(n=500)
print(df)

gdf_points = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df['LONGITUDE'], df['LATITUDE']))

world = gpd.read_file("https://naturalearth.s3.amazonaws.com/110m_cultural/ne_110m_admin_0_countries.zip")

usa = world[world['NAME'] == 'United States of America']

fig, ax = plt.subplots(figsize=(10, 6))
usa.plot(ax=ax, color='lightgray', edgecolor='black')
gdf_points.plot(ax=ax, color='dodgerblue', markersize=50, label='Points')
ax.scatter(-98.5795, 39.8283, color='red', marker='x', s=100, label='Center of the USA') 


plt.title('LAT, LONG over the USA map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
